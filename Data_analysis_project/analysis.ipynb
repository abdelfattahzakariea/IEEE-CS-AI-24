# Phase 2-apply the data analysis concepts

```python
# Remove columns with all null values
df_cleaned = data.dropna(axis=1, how='all')
df_cleaned.head(10)
```

```python
# casting date 
df_cleaned['Date of birth '] = pd.to_datetime(df_cleaned['Date of birth '],format='mixed')
#df_cleaned["Expected year of graduation"] = pd.to_datetime(df_cleaned['Expected year of graduation'],format='mixed',errors='coerce')

df_cleaned.head(10)
```

```python
df_cleaned.info()
```

```python
grouped = df_cleaned.groupby('City').count()
grouped
```

```python
df_cleaned['City'] = df_cleaned['City'].str.lower()
df_cleaned

def Cities_edits(value):
    if 'zag' in value or 'الز' in value:
        return 'Zagzig'
    if 'ram' in value:
        return '10 of Ramdan'
    if 'ibrahi' in value:
        return 'ibrahimya'
    if 'amh' in value or 'منيا' in value or 'umh' in value:
        return 'minyalqamh'
    if 'hammad' in value:
        return 'Abu Hammad'
    if 'abu k' in value or 'ابوكبير' in value :
        return 'Abu Kabir'
    if 'belb' in value:
        return 'belbias'
    if 'gm' in value or 'ديرب' in value:
        return 'diarb negm'
    if 'us' in value:
        return 'faqous'
    if 'mash' in value:
        return 'mashtoul elsouk elsharquiea'
    if 'mit' in value:
        return 'mit-ghamr'
    if 'sal' in value or 'الصالح' in value:
        return 'el salhia'
    if 'shar' in value or 'الشرقي' in value:
        return 'sharqiyea'
    if 'heh' in value or 'hahia' in value:
        return 'hehia'
    if 'fta' in value or 'زفت' in value:
        return 'Zefta'
    if 'cairo' in value or 'القاهره' in value:
        return 'cairo'
    if 'saqr' in value:
        return 'Kafr saqr'
    if 'mans' in value:
        return  'mansoura'
    else:
        return value
df_cleaned['City'] = df_cleaned['City'].apply(Cities_edits)
grouped = df_cleaned.groupby('City').count()
grouped.head(60)
```

```python
df_cleaned['Faculty'] = df_cleaned['Faculty'].str.lower()
df_cleaned

def faculty_edits(value):
    if 'comme' in value or 'comecra' in value:
        return 'commerce english'
    if 'engine' in value or 'هندس' in value:
        return 'engineering'
    if 'comput' in value or 'fci' in value or 'cse' in value:
        return 'computer and information science'
    if 'scie' in value:
        return 'science'
    if 'ture' in value:
        return 'agriculture'
    if 'art' in value:
        return 'faculty of arts'
    if 'scie' in value:
        return 'science'
    else:
        return value
    
df_cleaned['Faculty'] = df_cleaned['Faculty'].apply(faculty_edits)

grouped = df_cleaned.groupby('Faculty').count()
grouped.head(60)
```

```python
def excepted_year(value):
    if '2025-2026' in value:
        return '2026'
    if '2025/2026' in value or 'shaa' in value:
        return '2025'
    if 'Eng' in value or 'Gra' in value or 'very' in value:
        return 'gratuated'
    else:
        return value

df_cleaned['Expected year of graduation'] = df_cleaned['Expected year of graduation'].apply(excepted_year)

grouped = df_cleaned.groupby('Expected year of graduation').count()
grouped.head(60)
```

```python
def grade(value):
    if '4' in value:
        return '4th'
    if '3' in value:
        return '3rd'
    if '2' in value:
        return '2nd'
    else:
        return value
df_cleaned['Grade'] = df_cleaned['Grade'].apply(grade)
df_cleaned.tail(20)
```
